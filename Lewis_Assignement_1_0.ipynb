```
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        # Auto reject if the number is negative (because of the -) or if the number is above zero
        if x < 0 or (x%10 == 0 and x > 0): return False
        
        reverseX = 0
        #create the reverse of the right half of x by taking the digits off the back of x
        while x > reverseX:
            reverseX = reverseX * 10 + x % 10
            x = x // 10
        
        #cover odd and even numbers
        return x == reverseX // 10 or x == reverseX
```

````python
# Final working solution
class Solution(object):
   def isPalindrome(self, s):
        #where does each side's index start?
        string_start = 0
        string_end = len(s) - 1

        #examine left and right of string
        while (string_start < string_end):
            string_start_char = s[string_start]
            string_end_char = s[string_end]

            #are they not punctuation/spaces?
            #if they are, skip this one and ignore and move up an index
            if not string_end_char.isalnum():
                string_end -= 1
                continue
            if not string_start_char.isalnum():
                string_start += 1
                continue

            #do they actually match? if they dont, end early to say no.
            if not string_end_char.lower() == string_start_char.lower():
                return False

            #goto next indices
            string_end -= 1
            string_start += 1
        return True
````
